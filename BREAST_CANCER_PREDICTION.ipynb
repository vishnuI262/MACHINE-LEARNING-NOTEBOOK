```python
import gradio as gr
import numpy as np
import pandas as pd

# Importing the dataset
data = pd.read_csv("/content/BREAST CANCER.csv")
data = data.drop("Unnamed: 32", axis=1)
data = data.drop("id", axis=1)

from sklearn.preprocessing import LabelEncoder
data['diagnosis'] = LabelEncoder().fit_transform(data['diagnosis'])

# Selecting required columns
features = ['radius_mean', 'perimeter_mean', 'area_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'radius_worst', 'perimeter_worst', 'area_worst', 'compactness_worst', 'concavity_worst', 'concave points_worst']
x = data[features]
y = data["diagnosis"]

# Splitting the dataset
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=1)

# Training the Decision Tree model
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(xtrain, ytrain)

def func(RadiusMean, PerimeterMean, AreaMean, CompactnessMean, ConcavityMean, ConcavePointsMean, RadiusWorst, PerimeterWorst, AreaWorst, CompactnessWorst, ConcavityWorst, ConcavePointsWorst):
    inputs = np.array([RadiusMean, PerimeterMean, AreaMean, CompactnessMean, ConcavityMean, ConcavePointsMean, RadiusWorst, PerimeterWorst, AreaWorst, CompactnessWorst, ConcavityWorst, ConcavePointsWorst]).reshape(1, -1)
    prediction = model.predict(inputs)
    return "Malignant" if prediction[0] == 1 else "Benign"

# Creating the Gradio interface
interface = gr.Interface(func,
                         inputs=[gr.Textbox(label="RadiusMean"),
                                 gr.Textbox(label="PerimeterMean"),
                                 gr.Textbox(label="AreaMean"),
                                 gr.Textbox(label="CompactnessMean"),
                                 gr.Textbox(label="ConcavityMean"),
                                 gr.Textbox(label="ConcavePointsMean"),
                                 gr.Textbox(label="RadiusWorst"),
                                 gr.Textbox(label="PerimeterWorst"),
                                 gr.Textbox(label="AreaWorst"),
                                 gr.Textbox(label="CompactnessWorst"),
                                 gr.Textbox(label="ConcavityWorst"),
                                 gr.Textbox(label="ConcavePointsWorst")],
                         outputs=gr.Textbox(label="Diagnosis"),
                         title="Tumor type prediction using Decision Tree",
                         description="The Tumor type can be measured by the following metrics , and  can be classified as Benign or Malignant.")

# Launching the interface
interface.launch()
```
