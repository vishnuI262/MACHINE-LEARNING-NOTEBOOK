**IMPORTING THE NECESSARY LIBRARIES**

```python
import pandas as pd
```

```python
d=pd.read_csv("/content/Position_Salaries - Position_Salaries (1).csv")
```

```python
d.head()
```

**SPLITTING DEPENDENT AND INDEPENDENT VARIABLES**

```python
x=d.iloc[:,1:2]
y=d.iloc[:,2:]
```

**SPLITTING TRAINING AND TESTING DATA**

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
```

**FITTING LINEAR REGRESSION MODEL**

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred1=model.predict(xtest)
```

**R2_SCORE OF LINEAR REGRESSION MODEL**

```python
from sklearn.metrics import r2_score
r2_score(ytest,ypred1)
```

**PLOTTING OF LINEAR REGRESSION MODEL**

```python
import matplotlib.pyplot as plt
plt.scatter(xtrain,ytrain)
plt.plot(xtest,ypred1,color="black")
```

**MEAN SQUARED ERROR OF LINEAR REGRESSION MODEL**

```python
from sklearn.metrics import mean_squared_error as mse
mse(xtest,ypred1)
```

**Since , A Linear regression model is not capturing the relationship and the datapoints for the above dataset. We are changing the model to polynomial regression by adding polynomial features in the linear model.**

```python
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=2)
x_poly = pf.fit_transform(x)
model.fit(x_poly,y)
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,model.predict(x_poly))
```

Evaluating the degree of the polynomial.
When the degree is 2 , the model seems to be underfitted.

```python
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=3)
x_poly = pf.fit_transform(x)
model.fit(x_poly,y)
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,model.predict(x_poly))
plt.xlabel("LEVEL")
plt.ylabel("SALARY")
plt.title("SIMPLE POLYNOMIAL REGRESSION")
```

When the degree is 3 , the model seems to be good by covering the datapoints.

```python
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=4)
x_poly = pf.fit_transform(x)
model.fit(x_poly,y)
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,model.predict(x_poly))
```

When the degree is 4 , the model seems to be overfitted. So we are staying with the degree 3.

**FINAL OUTPUT**

```python
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=3)
x_poly = pf.fit_transform(x)
model.fit(x_poly,y)
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,model.predict(x_poly))
plt.xlabel("LEVEL")
plt.ylabel("SALARY")
plt.title("FINAL OUTPUT\nSIMPLE POLYNOMIAL REGRESSION")
```
