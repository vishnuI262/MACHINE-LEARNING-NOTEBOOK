 **IMPLEMENTATION OF GRADIENT DESCENT TO FIND THE BEST-FIT LINE FOR A LINEAR REGRESSION MODEL**

```python
import numpy as np
def gradient(x,y):
  m_curr=b_curr=0                                 # Initialize slope and intercept
  iterations=1000
  learning_rate=0.08
  y_p=[]
  n=len(x)
  for i in range(iterations):
    y_pred=m_curr*x+b_curr                 # Predicted values
    y_p.append(y_pred)

    # Calculate mean squared error and gradients
    l=(1/n)*sum([j**2 for j in (y-y_pred)])
    md=-(2/n)*sum(x*(y-y_pred))
    bd=-(2/n)*sum(y-y_pred)

    # Update slope and intercept using gradients and learning rate
    b_curr=b_curr-learning_rate*bd
    m_curr=m_curr-learning_rate*md
    print(m_curr,b_curr,l,i)
  return y_p

x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
d=gradient(x,y)
print(d)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
for i in d:
  plt.plot(x,i)
```
