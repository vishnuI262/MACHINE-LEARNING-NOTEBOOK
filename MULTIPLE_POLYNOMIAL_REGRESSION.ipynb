```python
import pandas as pd
```

```python
data=pd.read_csv("/content/houseprice - houseprice.csv")
```

```python
data["price"]=data["price"].fillna(data["price"].mean())
```

```python
data["bedrooms"]=data["bedrooms"].fillna(data["bedrooms"].mean())
```

```python
data["sqft_living"]=data["sqft_living"].fillna(data["sqft_living"].mean())
```

```python
#Replacing values
import numpy as np
data=data.replace('?',np.nan)
```

```python
data["bathrooms"]=data["bathrooms"].fillna(data["bathrooms"].mode)
```

```python
data=data.drop("Unnamed: 0",axis=1)
```

```python
data.corr()
```

```python
x=data[['sqft_living','grade']]
x
```

```python
y=data[['price']]
y
```

```python
from sklearn.preprocessing import PolynomialFeatures
P=PolynomialFeatures(degree=3)
xpoly=P.fit_transform(x)
```

```python
from sklearn.linear_model import LinearRegression
L=LinearRegression()
L.fit(xpoly,y)
```

```python
from sklearn.metrics import r2_score
r2_score(y,L.predict(xpoly))
```

```python
import matplotlib.pyplot as plt
x1_range=np.linspace(min(x.iloc[:,0]),max(x.iloc[:,0]),10)
x2_range=np.linspace(min(x.iloc[:,1]),max(x.iloc[:,1]),10)
x1_grid,x2_grid=np.meshgrid(x1_range,x2_range)
x_grid_poly=P.fit_transform(np.c_[x1_grid.ravel(),x2_grid.ravel()])
y_grid=L.predict(x_grid_poly).reshape(x1_grid.shape)
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x.iloc[:,0],x.iloc[:,1],y)
ax.plot_surface(x1_grid,x2_grid,y_grid)
plt.title("MULTIPLE POLYNOMIAL REGRESSION ")
```
