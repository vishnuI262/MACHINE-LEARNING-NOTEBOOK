```python
import h2o
```

```python
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

```python
import pandas as pd
data = pd.read_csv("/content/BREAST CANCER - BREAST CANCER.csv.csv")
```

```python
data=h2o.H2OFrame(data)
```

```python
data.head()
```

```python
y="diagnosis"
```

```python
train , test = data.split_frame(ratios=[0.8])
```

```python
automl = H2OAutoML(max_runtime_secs=600)
automl.train(y=y,training_frame=train)
```

```python
lb=automl.leaderboard
lb
```

```python
model=automl.leader
```

```python
d = model.model_performance(test)
print(d)
```
