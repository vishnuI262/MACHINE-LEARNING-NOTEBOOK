```python
import pandas as pd
data = pd.read_csv("/content/BREAST CANCER - BREAST CANCER.csv.csv")
```

```python
x = data.drop(columns=["diagnosis"])
y = data[["diagnosis"]]
```

```python
from sklearn.preprocessing import LabelEncoder
y = LabelEncoder().fit_transform(y)
```

```python
from tpot import TPOTClassifier
tpot = TPOTClassifier(generations = 7,scoring='neg_log_loss',population_size = 20,verbosity = 2,random_state = 62)
```

```python
from sklearn.model_selection import train_test_split
xtrain , xtest , ytrain , ytest = train_test_split(x,y,test_size=0.2,random_state=62)
```

```python
tpot.fit(xtrain , ytrain)
```

```python
accuracy = tpot.score(xtest , ytest)
accuracy
```

```python
ypred = tpot.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error as mse , roc_auc_score as score
import numpy as np
```

```python
mse = mse(ytest,ypred)
rmse=np.sqrt(mse)
auc_score = score(ytest,ypred)
gini = 2*auc_score-1
```

```python
print("MSE :", mse)
print("RMSE :", rmse)
print("AUC :", auc_score)
print("Gini Coefficient :", gini)
```
