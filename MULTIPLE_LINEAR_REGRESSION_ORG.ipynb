**IMPORTING THE NECESSARY LIBRARIES**

```python
import pandas as pd
import numpy as np
```

**IMPORTING THE DATASET**

```python
data=pd.read_csv("/content/drive/MyDrive/DATASET/startup-2 - startup-2 (1).csv")
```

**REMOVING UNNECESSARY COLUMNS**

```python
data=data.iloc[:,1:]
data
```

**FINDING NULL VALUES**

```python
data.isna().sum()
```

**IMPUTING NULL VALUES**

```python
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
data["State"]=imputer.fit_transform(data.iloc[:,3:4])
data
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
data["Administration"]=imputer.fit_transform(data.iloc[:,1:2])
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
data["R&D Spend"]=imputer.fit_transform(data.iloc[:,0:1])
```

```python
data.isna().sum()
```

**FILLING NULL VALUES BY AVERAGE**

```python
data["Profit"]=data["Profit"].fillna(data["Profit"].mean())
```

**ENCODING CATEGORICAL VALUES TO NUMERICAL VALUES**

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[3])],remainder='passthrough')
data=ct.fit_transform(data)
data=pd.DataFrame(data)
```

```python
data
```

**FEATURE SCALING - STANDARDIZATION**

```python
#Feature Scaling - Standardization

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data.iloc[:,3:5]=sc.fit_transform(data.iloc[:,3:5])
```

**FEATURE SCALING - NORMALIZATION**

```python
#Feature Scaling - Normalization

from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
data.iloc[:,5:7]=mms.fit_transform(data.iloc[:,5:7])
```

```python
data
```

**SPLITTING DEPENDENT AND INDEPENDENT VARIABLES**

```python
#Independent Variable [x]
x=data.iloc[:,0:6].values

#Dependent Variable [y]
y=data.iloc[:,-1].values
```

**SPLITTING TRAINING AND TESTING DATA**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

**FITTING LINEAR REGRESSION MODEL**

```python
from sklearn.linear_model import LinearRegression
l=LinearRegression()
l.fit(x_train,y_train)
```

**FINDING COEFFICIENTS AND INTERCEPT**

```python
l.coef_,l.intercept_
```

**PREDICTS TARGET VARIABLE USING TRAINING AND TESTING DATASET**

```python
ypred_test = l.predict(x_test)
ypred_train = l.predict(x_train)
```

**R2 SCORE FOR TRAINING AND TESTING DATASET**

```python
from sklearn.metrics import r2_score, mean_squared_error as mse
print("R2 SCORE (TEST) :",r2_score(y_test,ypred_test))
print("R2 SCORE (TRAIN) :",r2_score(y_train,ypred_train))
```

**MEAN SQUARED ERROR FOR TRAINING AND TESTING DATASET**

```python
print("MEAN SQUARED ERROR (TEST) :",mse(y_test,ypred_test))
print("MEAN SQUARED ERROR (TRAIN) :",mse(y_train,ypred_train))
```
